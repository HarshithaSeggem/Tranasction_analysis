```python
import pandas as pd

df = pd.read_csv("transactions (1).csv")
df['t_date'] = pd.to_datetime(df['t_date'])
```

```python
#1 What is the total sales amount across all transactions?
total_sales = df['t_amt'].sum()
total_sales
```

```python
#2 Which month had the highest total transaction amount?
df=df.groupby('t_date').sum().idxmax()
df
```

```python
#3What is the average transaction amount per customer?
avg_per_customer = df.groupby('cust_id')['t_amt'].mean()
avg_per_customer
```

```python
#4 What is the trend of total sales over months?
df['month'] = df['t_date'].dt.month_name()

monthly_trend = df.groupby('month')['t_amt'].sum()

monthly_trend
```

```python
#5 What is the highest single transaction amount recorded?
highest_transaction = df['t_amt'].max()
highest_transaction
```

```python
#6 Which service category contributes the most to total revenue?
services_revenue = df.groupby('services')['t_amt'].sum()
services_revenue.idxmax(), services_revenue.max()
```

```python
#7 Which product generated the highest revenue?
product_revenue = df.groupby('products_used')['t_amt'].sum()
product_revenue.idxmax(), product_revenue.max()
```

```python
#8 How does the average transaction amount vary between services?
avg_services_amt = df.groupby('services')['t_amt'].mean()
avg_services_amt

```
